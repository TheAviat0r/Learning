```python
import pandas as pd
import numpy as np
import wget
import requests

from __future__ import (absolute_import, division, print_function, unicode_literals)
```

```python
online_valid = wget.download('http://data.phishtank.com/data/online-valid.csv')
df = pd.read_csv('online-valid.csv', index_col='phish_id')
df = df.query('target == "Facebook"')
df.shape[0]
```

```python
df.head()
```

```python
df['submission_day'] = pd.to_datetime(df['submission_time']).dt.dayofyear
today_day = pd.to_datetime('today').dayofyear
df = df[today_day - df['submission_day'] <= 20]
```

```python
df.shape[0]
```

```python
working_urls = []

for url in df['url']:
    try:
        r = requests.get(url, timeout=10)
    except Exception:
        print('Timeout - ' + url)
    if r.status_code == 200:
        working_urls.append(url)
        
working_urls
```

```python
url_output = open('url_output.txt', 'w')

for url in working_urls:
    url_output.write(url + '\n')

url_output.close()

len(working_urls)
```
