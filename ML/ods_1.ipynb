```python
d = {'no' : False, 'yes' : True}
df['international plan'] = df['international plan'].map(d)
df['voice mail plan'] = df['voice mail plan'].map(d)
df.head()
```

```python
columns_to_show = ['total day minutes', 'total day calls', 'total day charge']
df.groupby(['churn'])[columns_to_show].describe(percentiles=[])
```

```python
df.groupby(['churn'])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

```python
pd.crosstab(df['churn'], df['voice mail plan'], normalize=True)
```

```python
df.pivot_table(['total day minutes', 'total day calls', 'total eve calls', 'total night calls'],
               ['area code'], aggfunc='mean').head()
```

```python
total_calls = df['total eve calls'] + df['total night calls'] + df['total day calls']
df.insert(loc=len(df.columns), column='total calls', value=total_calls)
```

```python
df['total charge'] = df['total day charge'] + df['total eve charge'] + df['total night charge']
df.head()
```

```python
pd.crosstab(df['churn'], df['international plan'], margins=True)
```

```python
pd.crosstab(df['churn'], df['customer service calls'], margins=True)
```

```python
df['many customers calls'] = (df['customer service calls'] > 3).astype('int64')
pd.crosstab(df['churn'], df['many customers calls'], margins=True)
```

```python
pd.crosstab((df['many customers calls'] & df['international plan']), df['churn'], margins=True)
```
