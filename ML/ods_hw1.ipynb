```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('hw1.csv', sep=';', index_col='id')
df.head()
```

```python
df.pivot_table(['height', 'weight'], ['gender'], aggfunc='mean')
```

```python
df.head()
```

```python
men = df[df['gender'] == 2]
women = df[df['gender'] == 1]
```

```python
smoking_men_percentage = men[men['smoke'] == 1].shape[0] / men.shape[0] * 100
smoking_women_percentage = women[women['smoke'] == 1].shape[0] / women.shape[0] * 100

print('smoking_men_percentage = %s' % (smoking_men_percentage))
print('smoking_women_percentage = %s' % (smoking_women_percentage))

diff = np.around(smoking_men_percentage / smoking_women_percentage)
if diff > 1:
    print('Men smoke more by %s%%' % (diff))
else:
    print('Women smoke more by %s%%' % (diff))
    
alco_men_percentage = men[men['alco'] == 1].shape[0] / men.shape[0] * 100;
alco_women_percentage = women[women['alco'] == 1].shape[0] / women.shape[0] * 100;

if alco_men_percentage > alco_women_percentage:
    print('Men consume more alcohol: %s%%' % (np.around(alco_men_percentage)))
else:
    print('Women consume more alcohol: %s%%' % (np.around(alco_women_percentage, decimals=1)))
```

```python
df['age_years'] = np.around(df['age'] / 365.0)
df['age'] = np.around(df['age'] / 30.0)
df.head()
```

```python
smoking_people = df[df['smoke'] == 1]
nonsmoking_people = df[df['smoke'] == 0]
smoking_age_median = smoking_people['age'].median()
nonsmoking_age_median = nonsmoking_people['age'].median()

nonsmoking_age_median - smoking_age_median
```

```python
table_people = df.query('age_years >= 60 and age_years <= 64 and smoke == 1')
table_people.head()
```

```python
chol_map = {1 : 4, 2 : 6, 3 : 8}
table_people['cholesterol'] = table_people['cholesterol'].map(chol_map)
table_people
```

```python
table_people.head()
```

```python
first_sample = table_people.query('ap_hi <= 120 and cholesterol == 4')
second_sample = table_people.query('(ap_hi > 160 and ap_hi < 180) and cholesterol == 8')
```

```python
first_share = first_sample.shape[0] / table_people.shape[0]
second_share = second_sample.shape[0] / table_people.shape[0]

np.around(first_share / second_share)
```

```python
df['height_m'] = df['height'] / 100.0
```

```python
df['bmi'] = df['weight'] / (df['height_m'] ** 2)
```

```python
women = df.query('gender == 1')
men = df.query('gender == 2')
if women['bmi'].mean() - men['bmi'].mean() > 0:
    print('women bmi is greater')
else:
    print('men bmi is greater')
```

```python
healthy = df.query('active == 1')
illy = df.query('active == 0')
if healthy['bmi'].mean() - illy['bmi'].mean():
    print('healthy bmi is greater')
else:
    print('illy bmi is greater')
```

```python
healthy_man = healthy.query('gender == 2 and alco == 0')
healthy_women = healthy.query('gender == 1 and alco == 0')
if healthy_man['bmi'].mean() - healthy_women['bmi'].mean() > 0:
    print('healthy man bmi is greater')
else:
    print('healthy women bmi is greater')
```
